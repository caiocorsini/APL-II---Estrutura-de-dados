```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("performace_data.csv")
df.head()
```

## Operação de Inserção

```python
df_insercao = df[df['OPERACAO'] == 'insercao']
df_insercao.head()
```

### Inserção de 10 itens - Tempo

```python
# tempo está em nanosegundos
df_inser_10 = df_insercao[df_insercao['N_ITENS'] == 10]
bst_data = df_inser_10[df_inser_10['TIPO_ARVORE'] == 'BST']
avl_data = df_inser_10[df_inser_10['TIPO_ARVORE'] == 'AVL']

width = 0.3
x = bst_data["ARVORE"]

fig, ax = plt.subplots(figsize=(20, 6))

# Plotar as barras para cada tipo de árvore
ax.bar(x - width/2, bst_data["MEDIA_TEMPO"], width=width, label="BST", color='skyblue')
ax.bar(x + width/2, avl_data["MEDIA_TEMPO"], width=width, label="AVL", color='salmon')

# Configurações dos eixos e título
ax.set_xlabel("ARVORE")
ax.set_ylabel("MÉDIA TEMPO (NS)")
ax.set_title("Média de Tempo de Inserção de 10 Itens")
ax.set_xticks(x)  # Define os valores de X como os números de 'ARVORE'
ax.legend(title="Tipo de Árvore")

plt.show()
```
